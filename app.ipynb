```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from IPython.display import display

# Function to plot the sine wave
def plot_sine(amplitude=1.0, frequency=1.0, phase=0.0):
    x = np.linspace(0, 2*np.pi, 500)
    y = amplitude * np.sin(2 * np.pi * frequency * x + phase)
    
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, color='blue')
    plt.ylim(-10, 10)
    plt.title(f"Sine Wave: amplitude={amplitude}, frequency={frequency}, phase={phase}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()

# Sliders for amplitude, frequency, and phase
amplitude_slider = FloatSlider(value=1.0, min=0, max=10, step=0.1, description='Amplitude')
frequency_slider = FloatSlider(value=1.0, min=0.1, max=5, step=0.1, description='Frequency')
phase_slider = FloatSlider(value=0.0, min=0, max=2*np.pi, step=0.1, description='Phase')

# Connect sliders to the plot function
interact(plot_sine, amplitude=amplitude_slider, frequency=frequency_slider, phase=phase_slider)
```
