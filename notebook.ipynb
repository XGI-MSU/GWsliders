# **Gravitational Wave Interactive Slider**

<div style="font-size: 17px;">

The Gravitational Wave Slider is an educational tool that lets you explore how parameter estimation works in binary black hole mergers. Move the sliders to match the original signal and reference parameters, and see how close you can get. The better the match, the more the waveform aligns with the data. The range of parameters that produce a reasonable waveform provides an uncertainty in your parameter estimation.

## **Parameters** <span style="font-size: 16px; font-style: italic;">Read more about the <a href="https://piper-morris.github.io/gw-resources/Gravitational_Wave_Parameter_Guide.pdf#view=FitV">parameters.</a></span>

**Sensitive Parameters:** 

* $m_{1}$: Mass of the first black hole (in $M_\odot$) 
* $m_{2}$: Mass of the second black hole (in $M_\odot$)               
* $\chi_{1}$: Dimensionless spin of the first black hole 
* $\chi_{2}$: Dimensionless spin of the second black hole 

**Insensitive Parameters:** 

* $\mathcal{M}$: Chirp mass of the system 
* $q$: Mass ratio of the system 
* $\chi_{+}$: Effective spin of the system 
* $\chi_{-}$: Mass-weighted spin difference of the system

**Other Variables:** 

* $Amplitude$: Strain amplitude 
* $Phase$: Phase at coalescence
* $\rho$: Signal-to-Noise Ratio (SNR)

## **Buttons**

<div style="line-height: 1.7;">

* **Reference Parameters:** Takes you to the published parameters for that waveform
* **Use Real Data:** Lets you choose from several real gravitational wave signals
* **Toggle Detectors:** Displays the signal from LIGO detectors in Hanford, Washington (H1) and Livingston, Louisiana (L1).
* **Residuals:** Plots the residuals [data-noise] of the signal
